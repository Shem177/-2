```python
# 1. Подключение нужных для работы моделей
import pandas as pd     # для чтения и записи табличных файлов
import numpy as np      # для математической обработки данных
import matplotlib.pyplot as plt # для построения графиков
from google.colab import data_table
data_table.enable_dataframe_formatter()
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_notebook()
from sklearn import metrics  
```

```python
# 2. Получение и чтение Excel файла из облака
print ('Сила тока, мкА')
url = "https://drive.google.com/uc?export=download&id=10X7i5PYwSAe08JtXUKDBkfWUHTuI8NMm"
input_df = pd.read_csv(url, sep = ';', header = None)
# вывод заголовка таблицы и первых пяти строк (сэмплов)
input_df.head()
```

```python
# 3. Просмотр последних строк таблицы
input_df.tail()
```

```python
# 4. Получение базовых статистических показателей
input_df.describe()
```

```python
# 5. Можно сразу построить графики
axes = input_df.plot(marker=None, figsize=(10, 10), subplots = True)
```

```python
#6 Перевод таблицы в матрицу 
input_matrix = input_df.values
print('размерность матрицы', input_matrix.shape)
print()

for i in range(1, len(input_df.columns)):
  print("{: >5} mean = {:.2f}".format(input_df.columns[i], input_matrix[0:2035 , i].mean()))
  
```

```python
#7 Расчет среднего зеначения 
A = input_df.iloc[0:1280 , 0]. values
print(A.shape)
abs(np.mean(A))

```

```python
#8 Создание функции мощности 
def P(A):
  return 10*A**2

  
```

```python
#9 Построение графика
fig, ax = plt.subplots(figsize = (20,10))
ax.plot(P(A))
ax.grid(True)
ax.set_xlabel('Сила тока, мкА')
ax.set_ylabel('Мощность,мкВт')
```

```python
#10 Построение интерактивного графика
p = figure(plot_width = 800, plot_height = 500)
p.line(np.arange(len(P(A))), P(A), line_width=2)
show(p)
```
